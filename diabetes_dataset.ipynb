```python
df =pd.read_csv("diabetics_data.csv")
```

# total_diabetes_patience

```python
ax =sns.countplot(data=df,x="diabetes")
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
dia_df =df[df["diabetes"]==1]
```

```python
dia_df.head()
```

# diabetes_in_male_and_female

```python
ax =sns.countplot(data=dia_df,x="gender")
for bars in ax.containers:
    ax.bar_label(bars)
```

# diabetes_in_age_group

```python
ax =sns.countplot(data=dia_df,x="age_group")
for bars in ax.containers:
    ax.bar_label(bars)
```

# diabetes_in_male_and_female_with_age_group


```python
ax =sns.countplot(data=dia_df,x="age_group",hue="gender")
for bars in ax.containers:
    ax.bar_label(bars)
```

# hypertension

```python
ax =sns.countplot(data=dia_df,x="hypertension")
for bars in ax.containers:
    ax.bar_label(bars)
```

# hypertension_in_male_female

```python
hyper =dia_df.groupby(["gender"],as_index=False)["hypertension"].sum().sort_values(by="hypertension",ascending=False)
sns.barplot(data=hyper,x='gender',y="hypertension")
```

# heart_disease

```python
ax =sns.countplot(data=dia_df,x="heart_disease")
for bars in ax.containers:
    ax.bar_label(bars)
```

# heart_disease_in_male_female

```python
heart =dia_df.groupby(["gender"],as_index=False)["heart_disease"].sum().sort_values(by="heart_disease",ascending=False)
sns.barplot(data=heart,x='gender',y="heart_disease")
```

# smoking_history

```python
ax =sns.countplot(data=dia_df,x="smoking_history")
for bars in ax.containers:
    ax.bar_label(bars)
```


# bmi_level_in_genders

```python
ax =sns.countplot(data=dia_df,x="bmi_level",hue="gender")
for bars in ax.containers:
    ax.bar_label(bars)
```

# hba1c_level_in_gender

```python
ax =sns.countplot(data=dia_df,x="hba1c_level",hue="gender")
for bars in ax.containers:
    ax.bar_label(bars)
```

# gulocose_level_in_blood_gender

```python
ax =sns.countplot(data=dia_df,x="bg_level",hue="gender")
for bars in ax.containers:
    ax.bar_label(bars)
```
